```python
from sklearn.datasets import make_classification
import numpy as np
```

```python
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=20)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100);
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]

```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
b
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input + b
x_input
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(solver='sag')
lor.fit(X,y)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

# solving the problem with perceptron by using sigmoid function


```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=sigmoid(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]

```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input2=np.linspace(-3,3,100)
y_input2=m*x_input + b
```

```python
plt.figure(figsize=(15,8))
plt.plot(x_input,y_input,color='red',linewidth=3,label='step')
plt.plot(x_input1,y_input1,color='black',linewidth=3,label='logistic_class')
plt.plot(x_input2,y_input2,color='yellow',linewidth=3,label='sigmoid')
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
plt.legend();
```

```python
def gd(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.5
    for i in range(1000):
        y_hat=sigmoid(np.dot(X,weights))
        weights=weights+ lr*(np.dot((y-y_hat),X)/X.shape[0])
    return weights[1:],weights[0]    
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
coef_,intercept_=gd(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m1=-(coef_[0]/coef_[1])
b1=-(intercept_/coef_[1])
```

```python
x_input3=np.linspace(-3,3,100)
y_input3=m1*x_input + b1
```
