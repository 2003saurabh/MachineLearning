```python
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split,RandomizedSearchCV
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
X = np.sort(10 * np.random.rand(100, 1) - 5, axis=0)  # Random values from -5 to 5
y = X**2 + np.random.randn(100, 1) * 2  # U-shaped quadratic relation with some noise
```

```python
plt.scatter(X,y)
```

```python
dt=tree.DecisionTreeRegressor(max_depth=2)
```

```python
dt.fit(X,y)
```

```python
tree.plot_tree(dt,rounded=True, 
    filled=True)
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
# Step 3: Plot decision surface (prediction values)
X_grid = np.arange(min(X), max(X), 0.01)  # Creating a grid of values
X_grid = X_grid.reshape((len(X_grid), 1))
y_pred = dt.predict(X_grid)

# Step 4: Visualize the decision tree regression surface
plt.scatter(X, y, color='red', label='Data points')
plt.plot(X_grid, y_pred, color='blue', label='Decision Tree Prediction')
plt.title('Decision Tree Regression on U-Shaped Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
